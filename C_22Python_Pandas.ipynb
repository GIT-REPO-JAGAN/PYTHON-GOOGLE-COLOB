#### Concatination

```python
data1 = {'ids': [101, 102, 103],
        'names': ['Alice', 'John', 'Jill'],
        'percents': [71.5, 86.5, 90.0]}
```

```python
data2 = {'ids': [201, 202, 203, 204],
        'names': ['Carol', 'Dan', 'John', 'Mark'],
        'percents': [87.5, 67.0, 91.5, 77.0]}
```

```python
df1 = pd.DataFrame(data1)
```

```python
df2 = pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis=0) # row-wise
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
students
```

```python
stu_df = pd.DataFrame(students)
```

```python
stu_df
```

```python
stu_df['percents']
```

```python
stu_df['percents'].sum()
```

```python
stu_df.drop([2], axis=0, inplace=True)
```
