#### Create a new file. Take the file name from the user
1. Write the contents in the file line by line
2. The contents should be taken as an user input (1 line at a time)
3. Keep taking the line input from the user until the user enters 'END'
4. Save the file

```python
x = input("Get the file name: ")
fh = open(x,'xt')
y = input("Get the file content: ")
while y != 'END':
    fh.write(y)
    fh.write('\n')
    y = input("Get the file content: ")
fh.close()
```

```python
file1=input('enter file name: ')
fh=open(file1,'x')
while True:
    l=input('enter content for line: ')
    if l=='END':
        break

    fh.write(l)
    fh.write('\n')

fh.close()
```

#### seek()  --> moves the file pointer to another position

```python
### fh.seek(offset, fromwhere)

# offset - number of bytes/characters
# fromwhere - 0,1,2
    # 0 => from the starting of the file
    # 1 => from the current position
    # 2 => from the end of the file
```

```python
fh = open('file1', 'tr')
```

```python
fh.tell()
```

#### Pickle   -->  Used to store structural data in a file (binary file)

```python
# also called as serialization
```

```python
# pickle module's dump() function
```

```python
import pickle
```

```python
emp_details = {'id': 12345, 'name':'John', 'dept': 'HR'}
```

```python
emp_details
```

```python
type(emp_details)
```

```python
fh = open('emp1.bin', 'xb')
pickle.dump(emp_details, fh)
fh.close()
```
