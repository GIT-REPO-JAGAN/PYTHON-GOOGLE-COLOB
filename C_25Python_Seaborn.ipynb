```python
ax = sns.countplot(x='PClass', data=titanic_df)

for patch in ax.patches:
    print(patch)
```

```python
ax = sns.countplot(x='PClass', data=titanic_df)

for patch in ax.patches:
    count = patch.get_height()
    x = patch.get_x()
    y = count
    coordinates = (x, y)
    ax.annotate(count, coordinates)
```

```python
ax = sns.countplot(x='PClass', data=titanic_df)

for patch in ax.patches:
    count = patch.get_height()
    x = patch.get_x() + 0.2
    y = count / 2
    coordinates = (x, y)
    ax.annotate(count, coordinates)
```

```python
ax = sns.countplot(x='PClass', data=titanic_df)

for patch in ax.patches:
    count = patch.get_height()
    x = patch.get_x() + 0.3
    y = count / 2
    coordinates = (x, y)
    ax.annotate(count, coordinates)
```
